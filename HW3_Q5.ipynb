**Rubidium - Strontium**

```
# Required Modules
import numpy as np
import matplotlib.pyplot as plt
```

```
# Rubidium Strontium Isotopic Composition
Sr87_Sr86 = np.array([0.7117, 0.7095, 0.7092, 0.7083, 0.7083, 0.7082])
Rb87_Sr86 = np.array([3.65, 1.8, 1.48, 0.82, 0.66, 0.74])
```

```
# Best fit line
line = np.polyfit(Rb87_Sr86, Sr87_Sr86, 1)
fit = np.poly1d (line)
dis_limit = np.linspace(0, 4, 100)
```

```
#Plot Figure

plt.figure(figsize = (8,6))
plt.plot(Rb87_Sr86, Sr87_Sr86, 'bo', label = 'Isotope Ratio')
plt.plot( dis_limit, fit(dis_limit),  'k-', label = 'Isochron')
plt.xlabel('Rb87/Sr86')
plt.ylabel('Sr87/Sr86')
plt.title('Rb - Sr Isochron')
plt.legend(loc='lower right')
plt.grid()
```
