**Argon - Argon**

```
# Required Modules
import numpy as np
import matplotlib.pyplot as plt
```

```
# Argon Ratios
Ar40_Ar36 = np.array([8855, 8439, 6867, 15380, 12970])
Ar39_Ar36 = np.array([1852, 1790, 1439, 3214, 2708])
```

```
# Best fit line
line = np.polyfit(Ar39_Ar36, Ar40_Ar36, 1)
fit = np.poly1d (line)
dis_limit = np.linspace(1400, 3300, 1900)
print(fit)
```

```
#Plot Figure

plt.figure(figsize = (8,6))
plt.plot(Ar39_Ar36, Ar40_Ar36, 'bo', label = 'Isotope Ratio')
plt.plot( dis_limit, fit(dis_limit),  'k-', label = 'Isochron')
plt.xlabel('Ar39/Ar36')
plt.ylabel('Ar40/Ar36')
plt.title('Ar40 - Ar39 Isochron')
plt.legend(loc='lower right')
plt.grid()

```
