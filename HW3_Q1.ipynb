**HW3 Question 1 -- 
 Evolution of Rabbit and Fox Population**


```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```
# Varibles for Rabbit 
R0 = 100      # Rabbit Population
B_r = 0.01    # Birth Rate
d_r = 0.005   # Death Rate
e_r = 0.0001

# Varibles for Fox
F0 = 100      # Fox population
B_f = 0.005   # Birth Rate
d_f = 0.01   # Death Rate
e_f = 0.0001

pop = [R0, F0]
```

**Rabbit Population Rate - dR/dt = βr*R − δr*R − εf*F*R**
 
**Fox Population Rate - dF/dt = βf*F − δf*F + εr*F*R.**

```
# Population evolution calculation
def dp_dt(I, t=0):
  return np.array([B_r*I[0] - d_r*I[0] - e_f*I[0]*I[1],
                  B_f*I[1] - d_f*I[1] + e_r*I[0]*I[1]])

time = np.arange(0, 6600, 1)   # Time vector in days
I = odeint(dp_dt, pop, time)   # Integrade over time scale
Rabbit, Fox = I.T
```

```
# Plotting the result
plt.figure(figsize = (8, 6))
plt.plot(time, Rabbit, 'b-', ms=2, label='Rabbit')
plt.plot(time, Fox, 'r-', ms=2, label='Fox')
plt.legend(loc='upper right')
plt.title('Population Evolution of the Rabbit and Fox')
plt.xlabel('Time (Days)')
plt.ylabel('Population')
```
