**Modules**

```
import numpy as np
import matplotlib.pyplot as plt
```

**Given Intensity (M), Declination (D) and Inclination (I)**

```
M = (1.0e-5)* np.array( [5.08, 3.87, 3.02, 2.1, 1.18] )
D = np.array([60.8, 62.1, 61.2, 62.2, 63.1])
I =  np.array([60.1, 59.8, 62.1, 61.6, 60.9])
```

**Computing East, North and Vertical Components**

```
N = np.asarray([i * np.cos(np.deg2rad(k)) * np.cos(np.deg2rad(j)) for i, j, k in zip(M, D, I)])
E = np.asarray([i * np.cos(np.deg2rad(k)) * np.sin(np.deg2rad(j)) for i, j, k in zip(M, D, I)])
V= np.asarray([i * np.sin(np.deg2rad(k)) for i, k in zip(M, I)])
```

**Printing N, E and V Components**

```
# Printing the three Coffecients as columns
print(np.c_[N, E, V])
```

**Question 3(b) - Linear FIt and Plot**

```
fit_1 = np.polyfit(E, N, 1)
line_1 = np.poly1d(fit_1)
x1 = np.linspace(0, np.amax(E), 100)

# Printing the linear fit for the data y = mx + c
print('Linear Fit Model Is', line_1)

# Plot 
plt.figure(figsize = (8,4))
plt.title('N Components Against E Components', fontsize=18)
plt.plot(E, N, 'or', label = 'N-E Data')
plt.plot( x1, line_1(x1), 'b', label = 'Linear Fit')
plt.xlabel('E Component (Am-1)', fontsize=14)
plt.ylabel('N Component (Am-1)', fontsize=14)
plt.tick_params(labelsize=12)
plt.legend(loc='lower right')
plt.grid()
```

**Question 3(c) - Linear FIt and Plot**

```
fit_2 = np.polyfit(E, V, 1)
line_2 = np.poly1d(fit_2)
x2 = np.linspace(0, np.amax(E), 100)

# Printing the linear fit for the data y = mx + c
print('Linear Fit Model Is', line_2)

# Plot 
plt.figure(figsize = (8,4))
plt.title('V Components Against E Components', fontsize=18)
plt.plot(E, V, 'or', label = 'N-E Data')
plt.plot( x2, line_2(x2), 'b', label = 'Linear Fit')
plt.xlabel('E Component (Am-1)', fontsize=14)
plt.ylabel('V Component (Am-1)', fontsize=14)
plt.tick_params(labelsize=12)
plt.legend(loc='lower right')
plt.grid()
```

```
fit_3 = np.polyfit(N, V, 1)
line_3 = np.poly1d(fit_3)
x3 = np.linspace(0, np.amax(N), 100)

# Printing the linear fit for the data y = mx + c
print('Linear Fit Model Is', line_3)

# Plot 
plt.figure(figsize = (8,4))
plt.title('V Components Against N Components', fontsize=18)
plt.plot(N, V, 'or', label = 'N-E Data')
plt.plot( x3, line_3(x3), 'b', label = 'Linear Fit')
plt.xlabel('N Component (Am-1)', fontsize=14)
plt.ylabel('V Component (Am-1)', fontsize=14)
plt.tick_params(labelsize=12)
plt.legend(loc='lower right')
plt.grid()
```
